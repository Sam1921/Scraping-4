```python
import pandas as pd
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
import time
```

```python
from bs4 import BeautifulSoup as BS
```

```python
driver = webdriver.Chrome('./chromedriver')
```

# Wikipedia

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
html = driver.page_source
```

```python
soup = BS(html, 'html.parser')
```

```python
table = soup.findAll("table",{"class":"wikitable sortable jquery-tablesorter"})[0]
```

```python
len(table.find("tbody").findAll("tr")) #Total video list
```

```python
video_df = pd.DataFrame({"Rank":[],"Name":[],"Artist":[],"Upload Date":[],"Views (billions)":[]})
```

```python
for row in table.find("tbody").findAll("tr"):
    video_df = video_df.append(pd.DataFrame({
                                "Rank": row.findAll("td")[0].text.replace(".",""),
                                "Name": row.findAll("td")[1].find("a").text,
                                "Artist": row.findAll("td")[2].text,
                                "Upload Date": row.findAll("td")[4].text,
                                "Views (billions)": row.findAll("td")[3].text
                                },index=[0]))
```

```python
video_df
```

# BCCI

```python
driver.get("https://www.bcci.tv/")
```

```python
#go to international fixture page
element_international = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]")
element_international.click()
time.sleep(10)
element_fixture = driver.find_element_by_xpath("/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a")
element_fixture.click()
```

```python
html = driver.page_source
```

```python
soup = BS(html, 'html.parser')
```

```python
bcci_df = pd.DataFrame({"Match Title":[],"Series":[],"Place":[],"Date":[],"Time":[]})
```

```python
for row in soup.find("div",{"class":"js-list"}).findAll("a"):
    bcci_df = bcci_df.append(pd.DataFrame({
                                            "Match Title":row.findAll("div")[3].findAll("strong")[1].text,
                                            "Series":row.findAll("div")[3].findAll("div")[0].findAll("span")[1].text,
                                            "Place":row.findAll("div")[3].findAll("div")[1].findAll("span")[1].text,
                                            "Date":row.findAll("div")[3].findAll("div")[1].find("strong").text,
                                            "Time":row.findAll("div")[1].find("span",{"class":"fixture__time"}).text
                                            },index=[0]))
```

```python
bcci_df
```

# Guru99

```python
driver.get("https://www.guru99.com/")
```

```python
# Search for "exception handling selenium"
element_search = driver.find_element_by_xpath("/html/body/div[2]/section[3]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/form/table/tbody/tr/td[1]/div/table/tbody/tr/td[1]/input")
element_search.send_keys('exception handling selenium')
time.sleep(10)
element_search_click = driver.find_element_by_xpath("/html/body/div[2]/section[3]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/form/table/tbody/tr/td[2]/button")
element_search_click.click()
time.sleep(20)
search_result = driver.find_element_by_xpath("/html/body/div[2]/section[3]/div/div/div/div/div/div/div/div[2]/div/div[1]/div/div/div/div[1]/div[6]/div[2]/div/div/div[1]/div[1]/div[1]/div[1]/div/a")
search_result.click()
time.sleep(10)
driver.switch_to.window(driver.window_handles[-1])
```

```python
html = driver.page_source
```

```python
soup = BS(html, 'html.parser')
```

```python
exception, description = [], []
```

```python
for row in soup.find("table",{"class":"table table-striped"}).findAll("tr"):
    exception.append(row.findAll("td")[0].text)
    description.append(row.findAll("td")[1].text)
```

```python
guru99_df = pd.DataFrame({"Exception":exception[1:],"Description":description[1:]})
```

```python
guru99_df
```

```python
gdp_df = pd.DataFrame({"Rank":[],
                       "State":[],
                       "GSDP(18-19)":[],
                       "GSDP(17-18)":[],
                       "Share(2017)":[],
                       "GDP($ billion)":[]
                      })
```
